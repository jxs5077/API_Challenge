### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations=city_data[["Latitude", "Longitude"]]

humidity=city_data["Humidity (%)"].astype(float)
```

```python
fig=gmaps.figure()

heat_layer=gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=10, point_radius=1)

fig.add_layer(heat_layer)

fig


#results in "This page can't load Google Maps correctly." error message from Google. g_key works, so not sure what the issue is.




```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
cities1= city_data[(city_data['Temperature (F)'] >70) & (city_data['Temperature (F)'] <80)]
cities2= cities1[(cities1["Wind (mph)"] < 10)]
cities3= cities2[(cities2["Cloudiness (%)"] == 0)]
cities3
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# search for lodging, not hotels, because it will include hospitals 

params = {
    "radius": 5000,
    "types": "lodging",
    "keyword": "lodging",
    "key": g_key}
    
for index, row in cities3.iterrows():
    
    lat = row["Latitude"]
    lng = row["Longitude"]
    
    params["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    hotel_df = requests.get(base_url, params=params)
    
    hotel_df = hotel_df.json()
    
    cities3.loc[index, "Hotel Name"] = hotel_df["results"][0]["name"]
    
cities3
```
